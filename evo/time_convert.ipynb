```python
number = "10"
f = open("./orbslam_v2" +number+ "_c.txt", "r")
f2 = open("./orbslam_v2" +number+ ".txt", "w")
first = False
first_time = None
for line in f:
    element = line.split()
    if (first == False):
        first = True
        first_time = float(element[0])
    time_convert = float(element[0]) - first_time
    print(time_convert)
    element[0] = "{:e}".format(time_convert)
    print(element[0] )
    element.append('\n')
    for i in element: 
        f2.write(i)
        f2.write(" ")
f2.close()
f.close()

```
